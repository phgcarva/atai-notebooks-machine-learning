```python
import matplotlib.pyplot as plt
```

This first function calculates the mean M(x) for a column x

```python
def mean(column):
    return sum(column)/len(column)
```

This next function calculates the variance of a column x according to the formula Var(x) = M(x²) - M(x)²

```python
def variance(column):
    return mean([x**2 for x in column]) - mean(column)**2
```

This function iterates through the columns of our dataset and finds the two with the greatest variances

```python
def find_two_columns(columns):
    variances = [(i,variance(column)) for i, column in enumerate(columns)]
    variances = sorted(variances, key=lambda x: x[1])
    return variances[-1][0], variances[-2][0]
```

Now, with these functions, we will load the dataset and find the two columns with the greatest variance

```python
with open('reality-check-data.txt') as f:
    lines = f.readlines()

data = [line.split("   ") for line in lines]

columns = []
for i in range(len(data[0])):
    column = [float(el[i]) for el in data]
    columns.append(column)

columns.pop(0)
first, second = find_two_columns(columns)

print("The greatest variance is in column {}".format(first+1))
print("The second greatest variance is in column {}".format(second+1))
```

And finally, we plot the data

```python
x = [line[first] for line in data]
y = [line[second] for line in data]

plt.plot(y,x, '.')
plt.show()
```
