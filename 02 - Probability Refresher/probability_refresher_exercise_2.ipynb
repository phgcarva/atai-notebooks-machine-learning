```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.tri as mtri
```

```python
def f(x,y):
    if x < y:
        return 6*(x**2 + y**2)
    else:
        return 0
```

```python
points = [(0.01*x, 0.01*y, f(0.01*x, 0.01*y)) for x in range(100) for y in range(100)]
```

```python
xs = np.array([x[0] for x in points])
ys = np.array([x[1] for x in points])
zs = np.array([x[2] for x in points])
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_trisurf(xs, ys, zs)
plt.show()
```

```python
def p_ygx(points, x):
    ys = [p[1] for p in points if p[0]==x]
    zs = [p[2] for p in points if p[0]==x]
    
    return ys, zs
```

```python
x_axis, y_axis = p_ygx(points, 0.2)

plt.plot(x_axis, y_axis)
plt.show()
```
